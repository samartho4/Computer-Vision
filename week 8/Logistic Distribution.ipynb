```python
age = [22,25,47,52,46,56,55,60,62,61,18,28,27,29,49]
```

```python
have_insurance = [0,0,1,0,1,1,0,1,1,1,0,0,0,0,1]
```

```python
len(age)
```

```python
len(have_insurance)
```

```python
import pandas as pd
```

```python
data = {"Age" : age , "have_insurance" : have_insurance}
```

```python
Data = pd.DataFrame(data)
```

```python
Data 
```

```python
X = Data.drop("have_insurance",axis =1 )
y = Data["have_insurance"]
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor
```

```python

model = regressor.fit(X,y) #actually produces the linear eqn for the data
model
```

```python
model.intercept_
```

```python
model.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
#plot for the TRAIN
 
plt.scatter(X,y, color='red') # plotting the observation line
 
plt.plot(X ,regressor.predict(X), color='blue') # plotting the regression line
 
 
plt.xlabel("Age") # adding the name of x-axis
plt.ylabel("Have insurance") # adding the name of y-axis
plt.show() 
```

```python
import math
```

```python
def sigmoid(p):
    return 1/ (1 + math.exp(-p))
```

```python
def prediction_function(age):
    z = model.intercept_  + model.coef_ * age
    q = sigmoid(z)
    return q
```

```python
age = int(input("enter your age"))
prediction_function(age)
if prediction_function(age) < 0.5:
    print( "person will not buy insurance")
else:
    print("person will buy insurance")
```
